```python
data = pd.read_csv("Line1_Section1.csv")
```

```python
StartX = data['E'][0]
StartY = data['N'][0]
data["Xdistance"] = np.sqrt((data["E"]-StartX)**2 + (data["N"]-StartY)**2)

data['D(m)'] =  data['D(m)']*-1

# cut out no data values
data2 = data[data['D(m)'] < 1000]
data3 = data[data['Resistivity_ohm-m'] < 1000]

fig, ax = plt.subplots(figsize =(15,7))
plt.scatter(data3['Xdistance'], data3['D(m)'], c=data3['Resistivity_ohm-m'], cmap=plt.cm.jet)
```
