# Produce some crossplots
Color codes by site, could use species though!

```python
Isla_list = [Maui_algae, Oahu_algae, Kauai_algae, BI_algae]
Title_list = ['Maui_algae', 'Oahu_algae', 'Kauai_algae', 'BI_algae']

# Parameter options are:  'ave_d15N', 'ave_d13C', 'ave_pctn', 'ave_pctc', 'std_d15N', 'std_d13C', 'std_pctn', 'std_pctc'  note BI does not have C or D13C
xP = "ave_d15N"
yP = "ave_pctn"

fig, ax = plt.subplots(2,2, figsize=(12, 12))

try: 
  for idx, Isla in enumerate(Isla_list):
    x = Isla[xP]
    y = Isla[yP]
    
    ax = plt.subplot(2,2,idx+1)
    g =sns.scatterplot(x=x, y=y, hue="Site", data=Isla, s=100, ax=ax).set_title("{} - {} vs. {}".format(str(Title_list[idx]), xP, yP))

    #    # plop on a regression line
    mask = ~np.isnan(x) & ~np.isnan(y) # regression does not like NaN values, this line masks them out and regression is only performed on finite numbers
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x[mask], y[mask])   # calculate regression stats for the ecoli and rainfall data
    r2 = r_value**2
    rX = np.linspace(min(x), max(x), 50)
    rY = slope*rX + intercept
    plt.plot(rX,rY, color='g', linestyle='--', alpha = .6, label = ("r$^2$ = " + "$%.2f$"%r2 +",\n Slope="+"$%.2f$"%slope +"\n std. Err="+"$%.2f$"%std_err) )
    plt.legend(loc=2) 

except: 
  pass
```

# Sweet. Interactive KML maps

```python
Isla_list = [Maui_algae, Oahu_algae, Kauai_algae, BI_algae]

# Parameter options are:  'ave_d15N', 'ave_d13C', 'ave_pctn', 'ave_pctc', 'std_d15N', 'std_d13C', 'std_pctn', 'std_pctc'  note BI does not have C or D13C
parameter = 'ave_d15N'   

# generate color list
green = Color("green")
colors = list(green.range_to(Color("red"),102))

# create the map
width, height = 750, 750
mapobject = folium.Map(location=[20.68568802, -156.444474], zoom_start=7, width=width, height=height)

# populate the map
for Isla in Isla_list:

  # ranges of parameter 
  p_max = Isla[parameter].max()                                                                        # used for scaling the symbols
  p_min = Isla[parameter].min()

  for index, row in Isla.iterrows():

      scaled_value = int(round(((row[parameter]-p_min)/(p_max-p_min)),2)*100)   # scale the value of the paramater for display
      color_RGB = colors[scaled_value].get_hex()

      folium.Marker(location=[row['Latitude'],row['Longitude']], popup=str(row), tooltip=row[parameter],
                    icon=folium.Icon(color='black', icon_color=color_RGB)).add_to(mapobject) 
mapobject
```
